```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from io import StringIO


data = pd.read_csv('data_log_transaction.csv')
```

```python
data['Transaction_Time'] = pd.to_datetime(data['Transaction_Time'])
data['Hour'] = data['Transaction_Time'].dt.hour
```

```python
data
```

```python
# 1. VOLUME TRANSAKSI
print("VOLUME TRANSAKSI")
volume = data['Device_Type'].value_counts()
total_transactions = len(data)
mobile_trans = (volume['Mobile'] / total_transactions * 100)
web_trans = (volume['Web'] / total_transactions * 100)

print(f"Mobile: {volume['Mobile']} transaksi ({mobile_trans:.2f}%)")
print(f"Web   : {volume['Web']} transaksi ({web_trans:.2f}%)")
print(f"â†’ Mobile {mobile_trans/web_trans:.2f}x lebih banyak dari Web")

```

```python
# 2. TOTAL NILAI TRANSAKSI
print("TOTAL NILAI TRANSAKSI")
amount_total = data.groupby('Device_Type')['Amount_IDR'].sum()
total_all = amount_total.sum()
percentage_mobile = (amount_total['Mobile'] / total_all * 100)
percentage_web = (amount_total['Web'] / total_all * 100)

print(f"Mobile: Rp {amount_total['Mobile']:,} ({percentage_mobile:.2f}%)")
print(f"Web   : Rp {amount_total['Web']:,} ({percentage_web:.2f}%)")
print(f"Nilai transaksi Mobile {amount_total['Mobile']/amount_total['Web']:.1f}x transaksi Web")
```

```python
# 3. SUCCESS RATE
print("SUCCESS RATE")
success_rate = data.groupby('Device_Type')['Status'].apply(
    lambda x: (x == 'Success').sum() / len(x) * 100
).round(1)

print(f"Mobile: {success_rate['Mobile']}%")
print(f"Web   : {success_rate['Web']}%")
print(f"â†’ Success rate Mobile {success_rate['Mobile']-success_rate['Web']:.1f}% dari Success rate Web")
```

```python
# 4. POLA WAKTU
print("POLA WAKTU TRANSAKSI")
# Mobile vs Web per jam
hourly_pattern = data.groupby(['Device_Type', 'Hour']).size().unstack(fill_value=0)

print("Mobile: Jam dengan transaksi terbanyak:")
mobile_peak = hourly_pattern.loc['Mobile'].nlargest(10)
for hour, count in mobile_peak.items():
    print(f"  Jam {hour}: {count} transaksi")

print("Web: Jam dengan transaksi terbanyak:")
web_peak = hourly_pattern.loc['Web'].nlargest(10)
for hour, count in web_peak.items():
    print(f"  Jam {hour}: {count} transaksi")
```

```python
# 5. USER LOYALTY ANALYSIS
print("USER LOYALTY")
user_device_pref = data.groupby('User_ID')['Device_Type'].apply(
    lambda x: x.value_counts().index[0] if len(x) > 1 else 'Mixed/Single'
)

loyal_users = user_device_pref.value_counts()
print(f"User yang konsisten pakai Mobile: {loyal_users.get('Mobile', 0)}")
print(f"User yang konsisten pakai Web   : {loyal_users.get('Web', 0)}")
print(f"User yang menggunakan keduanya   : {loyal_users.get('Mixed/Single', 0)}")

```

```python
# Cari heavy user (contoh: user dengan >5 transaksi)
heavy_users = data['User_ID'].value_counts()
heavy_user_example = heavy_users[heavy_users > 3].index[0] if len(heavy_users[heavy_users > 3]) > 0 else None

if heavy_user_example:
    heavy_user_data =data[data['User_ID'] == heavy_user_example]
    device_pref = heavy_user_data['Device_Type'].value_counts()
    print(f"\nContoh Heavy User ({heavy_user_example}):")
    print(f"  Total transaksi: {len(heavy_user_data)}")
    print(f"  Preferensi device: {device_pref.to_dict()}")
```

```python
# 6. RATA-RATA NILAI TRANSAKSI
print("RATA-RATA NILAI TRANSAKSI")
avg_amount = data.groupby('Device_Type')['Amount_IDR'].mean().round(0)
print(f"Mobile: Rp {avg_amount['Mobile']:,.0f} per transaksi")
print(f"Web   : Rp {avg_amount['Web']:,.0f} per transaksi")
```

```python
# 7. TIPE TRANSAKSI DOMINAN
print("DISTRIBUSI TIPE TRANSAKSI")
transaction_type_dist = pd.crosstab(data['Device_Type'], data['Transaction_Type'], normalize='index') * 100
print("Persentase per tipe transaksi:")
print(transaction_type_dist.round(1))
```

```python
print("\n" + "=" * 50)
print("ðŸŽ¯ KESIMPULAN UTAMA:")
print("=" * 50)
print(f"1. VOLUME: Mobile dominan ({mobile_trans:.1f}% vs {web_trans:.1f}%)")
print(f"2. NILAI: Mobile Rp {amount_total['Mobile']/1000000:.1f}M vs Web Rp {amount_total['Web']/1000000:.1f}M")
print(f"3. POLA: Mobile 24/7, Web jam kerja (9-17)")
print(f"4. RELIABILITY: Mobile lebih sukses (+{success_rate['Mobile']-success_rate['Web']:.1f}%)")
```

```python
from scipy import stats

# Hitung z-score untuk setiap tipe transaksi
data['Z_Score'] = 0
for tx_type in data['Transaction_Type'].unique():
    mask = data['Transaction_Type'] == tx_type
    data.loc[mask, 'Z_Score'] = np.abs(stats.zscore(data.loc[mask, 'Amount_IDR']))

# Buat figure
fig, axes = plt.subplots(1, 3, figsize=(18, 6))
fig.suptitle('ðŸŽ¨ SCATTER PLOT: Amount vs Transaction Time dengan Gradien Deviasi', 
             fontsize=16, fontweight='bold', y=1.02)

# Warna untuk setiap tipe transaksi
base_colors = {'Deposit': 'blue', 'Withdrawal': 'green', 'Fee': 'orange'}

for i, tx_type in enumerate(['Deposit', 'Withdrawal', 'Fee']):
    tx_data = data[data['Transaction_Type'] == tx_type]
    
    # Hitung statistik untuk garis referensi
    mean_amount = tx_data['Amount_IDR'].mean()
    median_amount = tx_data['Amount_IDR'].median()
    std_amount = tx_data['Amount_IDR'].std()
    
    # Scatter plot dengan gradien warna berdasarkan z-score
    scatter = axes[i].scatter(tx_data['Transaction_Time'], 
                             tx_data['Amount_IDR'], 
                             c=tx_data['Z_Score'],  # Color by z-score
                             cmap='RdYlBu_r',       # Red-Yellow-Blue reversed
                             s=70, 
                             alpha=0.7,
                             edgecolors='black', 
                             linewidth=0.5)
    
    # Tambah colorbar untuk setiap subplot
    cbar = plt.colorbar(scatter, ax=axes[i])
    cbar.set_label('Z-Score (Deviasi dari Rata-rata)', fontsize=9)
    
    # Garis referensi - RATA-RATA
    axes[i].axhline(y=mean_amount, color='red', linestyle='-', linewidth=2, 
                   label=f'Rata-rata: Rp {mean_amount:,.0f}')
    
    # Garis referensi - MEDIAN
    axes[i].axhline(y=median_amount, color='green', linestyle='--', linewidth=2, 
                   label=f'Median: Rp {median_amount:,.0f}')
    
    # Area Â±1 standard deviation dari mean (area normal)
    axes[i].axhspan(mean_amount - std_amount, mean_amount + std_amount, 
                   alpha=0.2, color='gray', label='Â±1 Std Dev')
    
    # Customize each subplot
    axes[i].set_title(f'{tx_type}\n({len(tx_data)} transaksi)', fontsize=14, fontweight='bold')
    axes[i].set_xlabel('Transaction Time')
    axes[i].set_ylabel('Amount_IDR')
    axes[i].yaxis.set_major_formatter(plt.FuncFormatter(lambda x, p: f'Rp {x:,.0f}'))
    axes[i].tick_params(axis='x', rotation=45)
    axes[i].legend(fontsize=9)
    axes[i].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

```python
# VERSI DENGAN COLOR CODING YANG LEBIH INFORMATIF
fig, axes = plt.subplots(1, 3, figsize=(20, 6))
fig.suptitle('ðŸ”¥ SCATTER PLOT: Anomali Detection dengan Risk Level Coloring', 
             fontsize=16, fontweight='bold', y=1.02)

for i, tx_type in enumerate(['Deposit', 'Withdrawal', 'Fee']):
    tx_data = data[data['Transaction_Type'] == tx_type].copy()
    
    # Hitung statistik
    mean_amount = tx_data['Amount_IDR'].mean()
    std_amount = tx_data['Amount_IDR'].std()
    Q1 = tx_data['Amount_IDR'].quantile(0.25)
    Q3 = tx_data['Amount_IDR'].quantile(0.75)
    IQR = Q3 - Q1
    
    # Kategorikan transaksi berdasarkan level risk/anomali
    def get_risk_level(amount):
        if amount <= Q1:
            return 0  # Low (normal)
        elif amount <= Q3:
            return 1  # Medium (normal)  
        elif amount <= Q3 + 1.5 * IQR:
            return 2  # High (mild outlier)
        elif amount <= Q3 + 3 * IQR:
            return 3  # Very High (outlier)
        else:
            return 4  # Extreme (severe outlier)
    
    tx_data['Risk_Level'] = tx_data['Amount_IDR'].apply(get_risk_level)
    risk_colors = ['green', 'limegreen', 'yellow', 'orange', 'red']
    risk_labels = ['Low', 'Medium', 'High', 'Very High', 'Extreme']
    
    # Plot dengan warna berdasarkan risk level
    for risk_level in sorted(tx_data['Risk_Level'].unique()):
        risk_data = tx_data[tx_data['Risk_Level'] == risk_level]
        axes[i].scatter(risk_data['Transaction_Time'], 
                       risk_data['Amount_IDR'], 
                       color=risk_colors[risk_level],
                       label=risk_labels[risk_level],
                       s=80, 
                       alpha=0.8,
                       edgecolors='black' if risk_level >= 3 else 'none',
                       linewidth=1.5 if risk_level >= 3 else 0.5)
    
    # Reference lines
    axes[i].axhline(y=mean_amount, color='blue', linestyle='-', linewidth=2, 
                   label=f'Mean: Rp {mean_amount:,.0f}')
    axes[i].axhline(y=Q1, color='gray', linestyle=':', linewidth=1, alpha=0.7)
    axes[i].axhline(y=Q3, color='gray', linestyle=':', linewidth=1, alpha=0.7)
    
    # Annotate beberapa transaksi extreme
    extreme_tx = tx_data[tx_data['Risk_Level'] == 4]
    for _, tx in extreme_tx.iterrows():
        axes[i].annotate(f"{tx['Transaction_ID']}\nRp {tx['Amount_IDR']:,}", 
                        xy=(tx['Transaction_Time'], tx['Amount_IDR']),
                        xytext=(tx['Transaction_Time'] + pd.Timedelta(hours=10), 
                               tx['Amount_IDR'] * 1.1),
                        arrowprops=dict(arrowstyle='->', color='red', alpha=0.7),
                        fontsize=8, color='red', ha='center')
    
    # Customize
    axes[i].set_title(f'{tx_type}\nRisk Level Distribution', fontsize=14, fontweight='bold')
    axes[i].set_xlabel('Transaction Time')
    axes[i].set_ylabel('Amount_IDR')
    axes[i].yaxis.set_major_formatter(plt.FuncFormatter(lambda x, p: f'Rp {x:,.0f}'))
    axes[i].tick_params(axis='x', rotation=45)
    axes[i].legend(title='Risk Level', title_fontsize=9, fontsize=8)
    axes[i].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

# SUMMARY STATISTICS
print("ðŸ“‹ SUMMARY ANOMALI DETECTION:")
print("=" * 60)
for tx_type in data['Transaction_Type'].unique():
    tx_data = data[data['Transaction_Type'] == tx_type]
    Q1 = tx_data['Amount_IDR'].quantile(0.25)
    Q3 = tx_data['Amount_IDR'].quantile(0.75)
    IQR = Q3 - Q1
    outlier_threshold = Q3 + 1.5 * IQR
    extreme_threshold = Q3 + 3 * IQR
    
    outliers = tx_data[tx_data['Amount_IDR'] > outlier_threshold]
    extreme_outliers = tx_data[tx_data['Amount_IDR'] > extreme_threshold]
    
    print(f"\n{tx_type.upper()}:")
    print(f"  â€¢ Total transaksi: {len(tx_data)}")
    print(f"  â€¢ Outliers (>Q3+1.5IQR): {len(outliers)} transaksi")
    print(f"  â€¢ Extreme outliers (>Q3+3IQR): {len(extreme_outliers)} transaksi")
    if len(extreme_outliers) > 0:
        print(f"  â€¢ Extreme outliers: {extreme_outliers[['Transaction_ID', 'Amount_IDR']].to_string(index=False)}")
```
